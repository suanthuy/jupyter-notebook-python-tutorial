# 1. Series Class<br>
* A class that deals with multiple indexed values<br>
    * Data structure similar to Python's list or dictionary data type.<br>
    * Create an object by receiving multiple values and indexes of each value.<br>
    * If you don't enter an index, 0-based index is used.

# 2. Creating Series Objects 

In[3]:

```python
import pandas as pd
```

```python
data = [10,5,87,32]
indices = ["Thomas", "Emily", "Sam", "Scott"]
```

In[4]:

```python
x = pd.Series(data) # Series from only data
```

In[5]:

```python
x
```

In[6]:

```python
x.index
```

In[7]:

```python
x = pd.Series(data, indices) # Series from data, and indices
```

In[8]:

```python
x
```

In[9]:

```python
x.index
```

In[12]:

```python
x.values
```

In[13]:

```python
x["Thomas"]
```

In[14]:

```python
x = pd.Series(data, index=indices, name="math_Score")
```

In[15]:

```python
x
```

In[16]:

```python
data_dict = {k:v for k, v in zip(indices, data)}
```

In[17]:

```python
data_dict
```

In[21]:

```python
x = pd.Series(data_dict)
```
