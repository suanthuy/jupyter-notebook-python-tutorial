# Reference by Array Slice<br>
* Array generated by slicing is "reference"<br>
    * Since slicing is a pass-by-reference operation, values are interlocked with each other.<br>
    * Explicitly copy the value to copy() method to create it as a pass-by-value.<br>
    * Indexing returns an immutable object, so it does not affect the original.

In[1]:

```python
import numpy as np
```

In[3]:

```python
x = np.arange(10)
slice = x[:4]
slice[2] = 100 # x[2] become 100
x
```

In[8]:

```python
x = np.arange(10)
x_copy = x[:4].copy()
x_copy[2] = 100 # x not changed
x_copy, x
```
